```python
c = r.TCanvas("", "", 1000, 500)
c.Divide(2, 1)

c.cd(1)
h_xz = reader_20230720_per0003.hist('z_xz:x_xz', '300, -16, 16, 300, -20, 30', cut_contidion='z_xz < -15')
h_xz.Draw("colz")

c.cd(2)
h_yz = reader_20230720_per0003.hist('z_yz:y_yz', '300, -16, 16, 300, -20, 30', cut_contidion='z_xz < -15')
h_yz.Draw("colz")

c.Draw()
```

```python
h = r.TH1D("", "", 100, -20, 30)
h.SetTitle("z_xz_ave (max(ev.z_xz) < -10);z [mm];")

for ev in reader_20230720_per0003.tree:
    if len(ev.z_xz) == 0:
        continue
    if max(ev.z_xz) < -10:
        continue

    z_xz_ave = sum(ev.z_xz) / len(ev.z_xz)
    h.Fill(z_xz_ave)

c = r.TCanvas("", "", 500, 500)
h.Draw()
c.Draw()
```

```python
h = r.TH1D("", "", 100, -20, 30)
h.SetTitle("z_yz_ave (max(ev.z_xz) < -10);z [mm];")

for ev in reader_20230720_per0003.tree:
    if len(ev.z_xz) == 0:
        continue
    if max(ev.z_xz) < -10:
        continue
    if len(ev.z_yz) == 0:
        continue

    z_yz_ave = sum(ev.z_yz) / len(ev.z_yz)
    h.Fill(z_yz_ave)

c = r.TCanvas("", "", 500, 500)
h.Draw()
c.Draw()
```

```python
h = r.TH1D("", "", 100, 0, 100)
h.SetTitle("length (max(ev.z_xz) < -10);z [mm];")

for ev in reader_20230720_per0003.tree:
    if len(ev.z_xz) == 0:
        continue
    if max(ev.z_xz) < -10:
        continue

    h.Fill(ev.length)

c = r.TCanvas("", "", 500, 500)
h.Draw()
c.Draw()
```
